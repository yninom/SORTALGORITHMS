1. Â¿Which algorithm is better? Plot them individually and all in one graph.
- $TA(n)=1000n$
- $TB(n)=200n\log{n}$
- $TC(n)=10n^{2}$
- $TD(n)=2^{n}$


```python
import numpy as np
import matplotlib.pyplot as plt
import math

n=np.arange(1e-7, 100, 0.1)
ta=1000*n
tb=200*n*np.log(n)
tc=10*np.power(n,2)
td=np.power(2,n)

figure, axis = plt.subplots(2, 2)
  
axis[0, 0].plot(n, ta)
axis[0, 0].set_title("TA")
axis[0, 1].plot(n, tb)
axis[0, 1].set_title("TB")
axis[1, 0].plot(n, tc)
axis[1, 0].set_title("TC")
axis[1, 1].plot(n, td)
axis[1, 1].set_title("TD")
  
plt.show()
```

```python
n=np.arange(1e-7, 10, 0.1)
ta=1000*n
tb=200*n*np.log(n)
tc=10*np.power(n,2)
td=np.power(2,n)
plt.plot(n,ta,'r',label='TA')
plt.plot(n,tb,'b',label='TB')
plt.plot(n,tc,'g',label='TC')
plt.plot(n,td,'k',label='TD')
plt.ylabel('$O(n)$: operations')
plt.xlabel('$n$: data points')
plt.legend()
plt.show()
```

In the previous graphs we can observe the complexity of the algorithms proposed as a function of the size of the input. TA represents a linear complexity that grows one thousand times proportionally with the size of the input, TB has log linear complexity, TC has quadratic complexity and TD has exponential complexity. For n between 0 and 10, the best algorithm is TD because it maintains a slow growth in the number of operations. For n between 10 and almost 100, the best algorithm is TC and for n greater than 100 the best algorithm is TA because it represents the slowest growth rate in the number of operations compared to the growth of the input and the constant 1000 doesn' matters.
